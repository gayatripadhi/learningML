```python
def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
```

```python
#Target variable (1 - client with payment difficulties: he/she had late payment more than X days on at least one of the first Y installments of the loan in our sample, 0 - all other cases)
tmp = application_train['TARGET'].value_counts()
total= application_train['TARGET'].count().sum()
df = pd.DataFrame({'labels': tmp.index,'values': (tmp.values/total)})
plt.figure(figsize = (6,6))
plt.title('Application Loans Replayed:Train')
sb.barplot(x ='labels', y='values', data=df)
plt.show()
#0 means loan is repayed,
#1 means loan is not repayed
```

```python
def plot_stats(feature,label_rotation=False,horizontal_layout=True):
    total= application_train[feature].count().sum()
    temp = application_train[feature].value_counts()
    df = pd.DataFrame({feature: temp.index,'Total number of contracts': (temp.values/total)})
    
    category_perc= application_train[[feature, 'TARGET']].groupby([feature],as_index=False).mean()
    
    if(horizontal_layout):
        fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16,6))
   
    s = sb.barplot(ax=ax1,x = feature, y="Total number of contracts",data=df)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    
    s = sb.barplot(ax=ax2,x = feature, y='TARGET', data=category_perc)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    plt.ylabel('Percent of target who have not repayed the loan')
    
    plt.tight_layout()
```

```python
def plot_distribution(feature):
    plt.figure(figsize=(10,6))
    plt.title("Distribution of %s" % feature)
    sb.distplot(application_train[feature].dropna(), kde=True,bins=100)
    plt.show()   
```

```python
#Income of the client
plot_distribution('AMT_INCOME_TOTAL')

#Credit amount of the loan
plot_distribution('AMT_CREDIT')

#Loan annuity
plot_distribution('AMT_ANNUITY')

#For consumer loans it is the price of the goods for which the loan is given
plot_distribution('AMT_GOODS_PRICE')
```
